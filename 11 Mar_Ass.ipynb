## Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

# Ans ==
1. T-test:
Use Case: The t-test is typically used when the sample size is small (usually less than 30) and the population standard deviation is unknown.
Example Scenario: Suppose you want to compare the mean heights of two groups of students (Group A and Group B) from two different schools. You collect a sample of 20 students from each group and measure their heights. Since the population standard deviation is unknown, you would use a t-test to determine if there is a statistically significant difference in the mean heights of the two groups.

2. Z-test:
Use Case: The z-test is appropriate when the sample size is large (typically when it exceeds 30) or when the population standard deviation is known.
Example Scenario: Imagine you are conducting a study to test the effectiveness of a new drug for a common medical condition. You have a large sample size (e.g., 1000 patients), and you know the population standard deviation of the response variable (e.g., blood pressure change) based on previous research. In this case, you can use a z-test to determine if the mean change in blood pressure is significantly different from zero, indicating the effectiveness of the drug.

## Q2: Differentiate between one-tailed and two-tailed tests.

# Ans==
The distinction between one-tailed (also known as one-sided) and two-tailed (also known as two-sided) tests relates to the directionality of the hypothesis being tested and where the critical region lies on the distribution of the test statistic.

1. One-Tailed Test:

Directionality: In a one-tailed test, the hypothesis specifies the direction of the effect (e.g., greater than or less than).

Critical Region: The critical region for the test is on one side of the distribution curve.

2. Two-Tailed Test:

Directionality: In a two-tailed test, the hypothesis does not specify the direction of the effect; it merely tests for the presence of an effect.

Critical Region: The critical region for the test is on both sides of the distribution curve.
