## Q2. Write a python program to print Python Exception Hierarchy.

```python
def print_exception_hierarchy():
    """
    Prints the Python Exception Hierarchy.
    """
    for exc in dir(__builtins__):
        if "Error" in exc:
            print(exc)
            try:
                for subexc in getattr(__builtins__, exc).__subclasses__():
                    print("    {}".format(subexc.__name__))
            except TypeError:
                pass
            print("\n")

if __name__ == "__main__":
    print_exception_hierarchy()
```

## Q5. Explain ImportError. What is ModuleNotFoundError?

## ANS ==
In Python, an ImportError is an error that occurs when the Python interpreter is unable to locate and load a module that is required by a program. This can happen for several reasons, including when the module name is misspelled, when the module file is missing or has been moved, or when the module requires another module that is not installed on the system.

ModuleNotFoundError, introduced in Python 3.6, is a subclass of ImportError. It is raised when the Python interpreter cannot find the module specified by the import statement. This error occurs when the module is not installed on the system, or when the module file is not present in the specified search path.

## Q6. List down some best practices for exception handling in python.

## ANS ==
1. Catch only the exceptions you expect: When writing exception handling code, it is important to catch only the specific exceptions that you expect to occur. Catching too many exceptions can make it harder to identify and debug issues.
2. Use try-except-finally blocks: Use try-except-finally blocks to handle exceptions. The try block contains the code that might raise an exception, the except block handles the exception if one occurs, and the finally block contains code that will always be executed, regardless of whether an exception occurred.
3. Provide helpful error messages: When raising an exception or handling an exception, provide an informative error message that describes the issue. This makes it easier for developers to debug issues.
4. Avoid using bare except clauses: Avoid using bare except clauses as they catch all exceptions, including system-level exceptions that could indicate a serious issue with the program or system.
5. Use built-in exceptions: Use built-in exceptions whenever possible, rather than creating your own custom exceptions. This makes it easier for other developers to understand your code and reduces the chance of naming conflicts.
