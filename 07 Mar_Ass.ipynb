## Q1. What are the three measures of central tendency?

#Ans ==
The three measures of central tendency are:

Mean: The mean, also known as the average, is calculated by adding up all the values in a data set and then dividing the sum by the total number of values. It is a common measure of central tendency.

Median: The median is the middle value of a data set when it is arranged in ascending or descending order. If there is an even number of values, the median is the average of the two middle values. The median is less sensitive to extreme values (outliers) than the mean.

Mode: The mode is the value that appears most frequently in a data set. A data set may have one mode (unimodal), more than one mode (multimodal), or no mode at all. Unlike the mean and median, the mode is not affected by extreme values.

These measures help provide a summary or central value that represents the typical or central tendency of a data set.

## Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?

#Ans ==
The mean, median, and mode are three different measures of central tendency, each serving as a way to summarize and describe the center of a dataset. Here are the key differences and uses for each:

Mean:

Definition: The mean is the average of all the values in a dataset. It is calculated by adding up all the values and dividing the sum by the total number of values.
Sensitivity to Outliers: The mean is sensitive to extreme values (outliers) in the dataset. A few extremely high or low values can significantly affect the mean.
Use: The mean is commonly used when the data is approximately symmetrically distributed and is not heavily skewed by outliers.
Median:

Definition: The median is the middle value of a dataset when it is arranged in ascending or descending order. If there is an even number of values, the median is the average of the two middle values.
Sensitivity to Outliers: The median is less sensitive to extreme values compared to the mean. It provides a better representation of the central tendency in the presence of outliers.
Use: The median is often preferred when the data is skewed or contains outliers. It is a robust measure that is not heavily influenced by extreme values.
Mode:

Definition: The mode is the value that occurs most frequently in a dataset. A dataset may have one mode (unimodal), more than one mode (multimodal), or no mode at all.
Sensitivity to Outliers: The mode is not affected by extreme values, as it is based on frequency counts of values rather than their actual magnitudes.
Use: The mode is useful for categorical or discrete data and can be informative about the most common values in a dataset. In some cases, a dataset may have no mode or multiple modes.

Skewness is a statistical measure that describes the asymmetry or lack of symmetry in a probability distribution or a dataset. In other words, skewness quantifies the degree and direction of skew (departure from horizontal symmetry) in the data distribution. A distribution or dataset can be positively skewed, negatively skewed, or approximately symmetric.

Here are the key characteristics of skewness:

Positively Skewed (Right Skewed):

The right tail of the distribution is longer or fatter than the left tail.
The majority of the data points are concentrated on the left side of the distribution.
The mean is typically greater than the median in a positively skewed distribution.

Negatively Skewed (Left Skewed):

The left tail of the distribution is longer or fatter than the right tail.
The majority of the data points are concentrated on the right side of the distribution.
The mean is typically less than the median in a negatively skewed distribution.

Symmetric (Zero Skewness):

The distribution is symmetric if the left and right sides are mirror images of each other.
The mean and median are equal in a symmetric distribution.
The skewness value is close to zero.

Skewness is a valuable tool in descriptive statistics as it provides insights into the shape and nature of a dataset. It helps analysts and researchers understand the distribution's departure from symmetry and can influence the choice of appropriate statistical measures for central tendency and dispersion. Skewness is often calculated using mathematical formulas, and software packages can automatically compute it for a given dataset.

## Q12. For a normal distribution data what is the relationship between its measure of central tendency?

# Ans ==
For a normal distribution, there is a specific relationship among its measures of central tendency: the mean, median, and mode. In a normal distribution, these measures are all equal and located at the center of the distribution. Here's the key relationship:

Mean (μ):

In a normal distribution, the mean (μ) is located at the center of the distribution.

Median:

Mode:

## Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

# Ans ==

Outliers can significantly impact measures of central tendency and dispersion in a dataset. Here's how outliers affect these measures:

Measures of Central Tendency:

Mean:

Effect: Outliers can heavily influence the mean because it involves summing up all values and dividing by the number of values.
Example: Consider the dataset [10, 15, 20, 25, 100]. The mean without the outlier is (10 + 15 + 20 + 25) / 4 = 17.5. With the outlier, the mean becomes (10 + 15 + 20 + 25 + 100) / 5 = 34.
Median:

Effect: The median is less affected by outliers since it is the middle value. However, extreme values can still shift the median.
Example: In the dataset [10, 15, 20, 25, 100], the median without the outlier is 20. With the outlier, the median becomes 25.
Mode:

Effect: Outliers usually have little effect on the mode, as it represents the most frequent value(s).
Example: In the dataset [10, 15, 20, 25, 100], the mode remains the same (no mode in this case) regardless of the outlier.
Measures of Dispersion:

Range:

Effect: Outliers can significantly affect the range, as it is the difference between the maximum and minimum values.
Example: In the dataset [10, 15, 20, 25, 100], the range without the outlier is 100 - 10 = 90. With the outlier, the range becomes 100 - 10 = 90.
Variance and Standard Deviation:

Effect: Outliers can substantially increase the variance and standard deviation since these measures involve the squared differences from the mean.
Example: Using the dataset [10, 15, 20, 25, 100], the standard deviation without the outlier is relatively low. With the outlier, the standard deviation increases significantly due to the influence of the large difference from the mean.
