## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

# Ans == 
1.Line Plot:
Use: Visualize the relationship between two continuous variables over a continuous interval. It's useful for showing trends over time.
2.Scatter Plot:
Use: Display the relationship between two numerical variables. Helpful for identifying patterns and outliers.
3.Bar Plot:
Use: Compare categorical data, showing the distribution of a single variable for different categories.
4.Box Plot (Box-and-Whisker Plot):
Use: Display the distribution of a dataset and identify outliers. Useful for comparing distributions.
5.Heatmap:
Use: Show the correlation between different variables in a dataset. Particularly helpful in identifying patterns and relationships in large datasets.


## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
fmri_data = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
#Ans ==
import seaborn as sns
import matplotlib.pyplot as plt
titanic_data = sns.load_dataset("titanic")
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
sns.boxplot(x='pclass', y='age', data=titanic_data, ax=axes[0])
axes[0].set_title('Box Plot of Age by Pclass')
sns.boxplot(x='pclass', y='fare', data=titanic_data, ax=axes[1])
axes[1].set_title('Box Plot of Fare by Pclass')
plt.tight_layout()
plt.show()

```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
#Ans ==
diamonds_data = sns.load_dataset("diamonds")
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple="stack")
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
#Ans ==
iris_data = sns.load_dataset("iris")
sns.pairplot(iris_data, hue="species")
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
#Ans ==
# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Reshape the data for heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plot a heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu")
```
